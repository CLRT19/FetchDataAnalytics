```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dictionary = {1:4222, 2:3830, 3:4767, 4:4882, 5:5627, 6:5405, 7:6058, 8:6191, 9:6355, 10:7305, 11:7512, 12:8447}

plt.bar(list(dictionary.keys()), dictionary.values())
plt.show()
```

```python
dictionary2 = {1:97, 2:87, 3:89, 4:90, 5:88, 6:88, 7:88, 8:88, 9:88, 10:91, 11:93, 12:98}

plt.bar(list(dictionary2.keys()), dictionary2.values())
plt.show()
```

```python
mean1 = np.mean(list(dictionary.values()))
print('mean', mean1)
std1 = np.std(list(dictionary.values()))
print('standard deviation', std1)
mean2 = np.mean(list(dictionary2.values()))
std2 = np.std(list(dictionary2.values()))
print('mean',mean2)
print('standard deviation',std2)
```

```python
a1 = np.array(list(dictionary.values()))
a2 = np.array(list(dictionary2.values()))
a3 = a1/a2
print(a3)
```

```python
plt.bar(list(dictionary2.keys()), a3)
```

```python
mean3 = np.mean(a3)
std3 = np.std(a3)
print('mean',mean3)
print('standard deviation',std3)
```
